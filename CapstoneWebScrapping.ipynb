```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
#Extract the product names
names= [i.text for i in soup.find_all('h3',class_="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2")]
#Extract the number of colors the product is available in
colors = [i.text.strip('C')[0] for i in soup.find_all('div',class_="relative h-12 mobile:h-10 flex items-center group")]

#extracting the price of each product and the category
prices = []
category = []
for i in soup.find_all('div',class_="flex flex-col flex-none mobile:items-start items-end text-sm md:text-base mobile:mt-2"):
    price_text = i.text
    match = re.search(r'\$\d+.\d+', price_text)
    if match:
        prices.append(match.group())
        category.append("Men")  
```

```python
#Extracting the reviews
reviews = []
for div in soup.find_all('div', class_="relative w-full flex flex-col gap-2"):
    # Check if the desired class exists within the current div
    class_exists = div.find(class_="ml-1 pl-px text-sm")
    
    # If the class exists, append its text to the result list
    if class_exists:
        reviews.append(class_exists.text.strip("Reviews"))
    else:
        # If the class doesn't exist, append "-"
        reviews.append("-")
```

```python
#Extracting the ratings
ratings = []
for div in soup.find_all('div', class_="relative w-full flex flex-col gap-2"):
    class_exists = div.find(class_="flex items-center")
    if class_exists:
        ratings_text = class_exists.text
        match = re.search(r'\b\d+\.\d+\b', class_exists.text)
        if match:
            ratings.append(float(match.group()))
        
    else:
        # If the class doesn't exist, append "-"
        ratings.append("-")
```
